```python

import asyncio
import aiohttp
from synology_dsm import SynologyDSM

async def main():
    print("Creating Valid API")
    async with aiohttp.ClientSession(
        connector=aiohttp.TCPConnector()
    ) as session:
        await do(session)

async def do(session: aiohttp.ClientSession):
    api = SynologyDSM(session, "naschauray.synology.me", "5001", "valentin", "bonjourjesuisSpiderman", debugmode=True)
    await api.login()

    print("=== Video ===")
    print("=== Devices ===")
    devices = await api.video_station.get_devices()
    for device in devices:
        print(device.device_id)
        print(device.device_title)
    print("=== Movies ===")
    movies = await api.video_station.get_list_movie(0, 2, True)
    for movie in movies:
        print(f"'{movie.movie_id}' has {movie.title}")
        print(f"'{movie.summary}'")
        print(f"'{movie.poster_url}'")
        print(f"'{movie.file_id}'")
        print(f"'{movie.file_name}'")
        print(f"'{movie.file_path}'")
        print(f"'{movie.file_video_codec}'")
        print(f"'{movie.file_audio_codec}'")
        # print(api.video_station.get_url_poster(movie))
        metadata = await api.video_station.get_metadata_movie(movie.movie_id)
        print(metadata)

await main()
```
